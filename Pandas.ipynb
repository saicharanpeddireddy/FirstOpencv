```python
 import pandas as ps
 df3=ps.read_excel('/content/drive/MyDrive/Sample Input.xlsx',skiprows=[0])
```

```python
display(df3.head(1))
```

```python
#Unpivot the date columns
result_melt=df3.melt(id_vars=['DCD:M:120','Description','Frequency','Units','Region'], var_name='Date', value_name='VALUE')
```

```python
#Result2=result1.pivot(['DCD:M:120','Description','Frequency','Region','Date'],columns='Units')
# ignoring the 'DCD:M:120','Description' columns Because of dupilicates
piv=result_melt.pivot(['Frequency','Region','Date'],columns='Units',values='VALUE')
display(piv)
```

```python
piv.columns = [''.join(str(s).strip() for s in col if s) for col in piv.columns]

#reset index
piv.reset_index(inplace=True)

piv.rename(columns = {'VALUEIndex':'Index', 'VALUEMw':'Mw','VALUEPer cent':'Per cent'}, inplace = True)
#view updated pivot table
print(piv.head(2))

#to remove charecters after COMMA in Region column
piv['Region']=piv['Region'].apply(lambda j:j.replace(j[j.index(','):],''))
```
